```python
down_payment = int(input("Enter the down payment amount: "))
payment_per_month = int(input("Enter the payment per month amount: "))
num_months = int(input("Enter the number of months: ")) 

total_cost = down_payment + (payment_per_month * num_months)

print("The total cost of the item is:", total_cost)
```
