There are N candidates standing in a line.
Each candidate has a score written on their badge.

You are given:

An integer N

An array of N integers representing scores of candidates.

ðŸ“Œ Task:

Count the number of candidates whose score is strictly greater than the average score of all candidates.

```python
n = int(input())
arr = list(map(int, input().split()))
count = 0
avg = sum(arr)//n
for score in arr:
    if score > avg:
        count +=1

print("Number of Subjects :",n)
print('Marks         :' ,arr)
print("Average marks :",avg)
print("Avg marks peksha kmi asnare students chi sankhya : ",count)
```

You are given:

An integer N

A list of N integers

ðŸ“Œ Task:

Find the second largest distinct element in the list.

If it does not exist, print:

No second largest

```python
n = int(input())
numbers = list(map(int, input().split()))
print(f'Your Input :{n}  {numbers}')

if n < 2:
    print("No second largest")
else:
    largest = numbers[0]
    second_largest = None

    for value in numbers[1:]:
        if value > largest:
            second_largest = largest
            largest = value
        elif value != largest:
            if second_largest is None or value > second_largest:
                second_largest = value

    if second_largest is None:
        print("No second largest")
    else:
        print(second_largest)

```
