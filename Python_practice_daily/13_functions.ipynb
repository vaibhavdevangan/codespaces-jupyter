```python
def calc_pizza_volume(pizza_diameter, pizza_height):
    pi_val = 3.14159265
    
    pizza_radius = pizza_diameter / 2.0
    pizza_area = pi_val * pizza_radius * pizza_radius
    pizza_volume = pizza_area * pizza_height
    return pizza_volume
```

#### Program with a function to convert height in feet/inches to centimeters.

```python
CM_PER_INCH = 2.54
INCHES_PER_FOOT = 12

def height_US_to_cm(feet, inches):
    total_inches = feet*INCHES_PER_FOOT + inches
    cm = total_inches * CM_PER_INCH
    return cm

i_feet = 6
i_inches = 4

centimeters = height_US_to_cm(i_feet, i_inches)
print('Centimeters:',centimeters)
```

### Arbitrary numbers of position arguments using *args.


```python
def print_sandwich(bread, slice, *args):
    print(f' Arbitary arguments are:' ,*args)
    print(f'{slice} on {bread}', end = " ")
    if len(args) > 0:
        print( 'with' , end= " ")
    for extra in args:
        print(extra, end = " ")

print_sandwich( 'sourdough' ,'kabab', 'mayo')
```
