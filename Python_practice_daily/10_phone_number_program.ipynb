```python

user_input = input( 'Enter phone number: ')
phone_number = ''

for character in user_input:
    # print(f' Element {index} is: {character}')
    if character.isdigit() or character == '-'  :
        phone_number += character
    elif character.upper() in ['A', 'B', 'C'] or character.lower() in ['a', 'b', 'c']:
        phone_number += '2'
    elif character.upper() in ['D', 'E', 'F'] or character.lower() in ['d', 'e', 'f']:
        phone_number += '3'
    elif character.upper() in ['G', 'H', 'I'] or character.lower() in ['g', 'h', 'i']:
        phone_number += '4'
    elif character.upper() in ['J', 'K', 'L'] or character.lower() in ['j', 'k', 'l']:
        phone_number += '5'
    elif character.upper() in ['M', 'N', 'O'] or character.lower() in ['m', 'n', 'o']:
        phone_number += '6'
    elif character.upper() in ['P', 'Q', 'R', 'S'] or   character.lower() in ['p', 'q', 'r', 's']:
        phone_number += '7'
    elif character.upper() in ['T', 'U', 'V'] or character.lower() in ['t', 'u', 'v']:
        phone_number += '8'
    elif character.upper() in ['W', 'X', 'Y', 'Z'] or character.lower() in ['w', 'x', 'y', 'z']:
        phone_number += '9'
    else:
        phone_number += '?'

print(f' phone number is: {phone_number}')
```
