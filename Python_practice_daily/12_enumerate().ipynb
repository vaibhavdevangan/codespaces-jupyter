##### Using range() and len() to iterate over a sequence.


```python
origins = ['Earth', 'Mars', 'Venus', 'Jupiter', 'Saturn']
for index in range(len(origins)):
    value = origins[index] #Retrieve value of element
    print(f'Element {index}: {value}')
```

##### 
Using enumerate() to find the index of each element.

```python
origins ="vaibhav"
for (index, value) in enumerate(origins):
    print(f'Element {index}: {value}')

# Using enumerate to get index and value 
# can be run on any iterable object like list, tuple, string etc.
# Here, we used it on a string 'origins'

```
