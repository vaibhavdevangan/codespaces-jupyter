What is numpy?

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array
object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays,
including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic
statistical operations, random simulation and much more.

```python
import numpy as np  #importing numpy library for numerical operations


# 1D array # Vector
a = np.array([1, 2, 3])  #creating a numpy array with elements 1, 2, and 3
```

### Array Attributes

```python
a1 = np.arange(1, 10)  #Vector
a2 = np.arange(12, dtype=float).reshape(3,4)  #Matrix
a3 = np.arange(24).reshape(2, 3, 4)  #Tensor
```

```python
print("Vector a1:\n", a1)  #printing the 1D array (vector)
print()
print("Matrix a2:\n", a2) #printing the 2D array (matrix)  
print()
print("Tensor a3:\n", a3)  #printing the 3D array (tensor)
```

```python
#ndim  #checking the number of dimensions of a numpy array
print("Number of dimensions in a1:", a1.ndim)  #checking the number
print("Number of dimensions in a2:", a2.ndim)  #checking the number of dimensions in the 2D array (matrix)
print("Number of dimensions in a3:", a3.ndim)  #checking the number of dimensions in the 3D array (tensor)
```

```python
#shape  #checking the shape of a numpy array
print("Shape of a1:", a1.shape)  #checking the shape of the 1D array (vector)
print("Shape of a2:", a2.shape)  #checking the shape of the 2D array (matrix)
print("Shape of a3:", a3.shape)  #checking the shape of the 3D array (tensor)
```
