```python
prices = {"apple": 0.5, "banana": 0.3, "orange": 0.8}   
print(f'Fruit Prices: {"apple"} costs ${prices["apple"] }'  )
```

```python
prices = {}
prices["apple"] = 0.5 # Adding key-value pair
print(prices["apple"])
```

```python
prices["apple"] = 0.6 # Updating value for existing key
print(prices["apple"])
```

```python
del prices["banana"]  # Deleting a key-value pair
print(prices)
```
