```python
Iterating over a list using a for loop.

1) The first iteration assigns the variable name with 'Bharat' and prints 'Hi Bharat!' to the screen.
```

```python

for name in ["Bharat","Nutan","Jay"]:
    print(f'Hi {name}!')

```

A for loop assigns the loop variable with a dictionary's keys.

```python

channels = {
'MTV': 35,
'CNN': 28,
'FOX': 11,
'NBC': 4,
'CBS': 12

}

for c in channels:
    print(f'{c} is on channel {channels[c]}')
```

For loop example: Calculating shop revenue.

```python

daily_revenues = [
2356.23,            # Monday
1800.12,
1792.50,
2058.10,
1988.00,
2002.99,
1890.75             # Sunday

]




total = 0
for day in daily_revenues:
    total += day

average = total / len(daily_revenues)

print(f'Weekly revenue: ${total :.2f}')
print(f'Daily average revenue: ${average :.2f}')
```

For loop example: Looping over a sequence in reverse.

```python
names = [
'Ajay',
'Bharat',
'Chetan',
'Durgesh',
'Ethan'
]

for name in names:
    print(name,' |',end=' ')

print("\nPrinting in reverse:")
for name in reversed(names):
    print(name,'|',end=' ')
```
