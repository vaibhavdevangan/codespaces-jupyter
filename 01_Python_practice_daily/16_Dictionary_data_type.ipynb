### Dictionaries

```python
# 2. Create and Print Dictionary:

# Creating a dictionary
my_dict = {'apple' : 5, 'banana': 3, 'orange': 7}

# Printing the dictionary
print(my_dict)
```

```python
# 2) Indexing operation: retrieves the value associated with key.

# syntax: my_dict[key]

a = my_dict['apple']
print(a)
```

```python
# 3) Addition of key and value: Adds an entry if the entry does not exist, else modifies the existing entry.

# syntax: my_dict[key] = value

my_dict['mango' ]=8

print(my_dict)
```

```python
# 4) Dictionary update() method: updates the dictionary with the elements from another dictionary object or from an iterable of key/value pair

# syntax: my_dict.update([other])

print(my_dict)

{'apple': 10, 'banana': 3, 'orange': 7, 'mango': 8}

my_dict.update(apple=15,mango=20)

my_dict
```

```python
# 5) Remove a Key from a Dict Using the 'del' Keyword:Deletes the key from a dict.
# syntax: del my_dict[key]

print(my_dict)

{'apple': 15, 'banana': 3, 'orange': 7, 'mango': 20}

del my_dict['banana']

print(my_dict)
```

```python
# 6) Using the 'in' Operator to Test for existence of "key" in "my_dict": use the in operator to check if a key exists in a dictionary.

# syntax: key in my_dict

if 'apple' in my_dict:
    print("Key exists in the dictionary.")
else:
    print("Key does not exist in the dictionary.")

```


Dictionary methods

```python
my_dict = {'Ahmad': 1, 'Jane': 42}
my_dict.clear()
print(my_dict)

my_dict = {'Ahmad': 1, 'Jane': 42}
print(my_dict.get('Jane', 'N/A'))
print(my_dict.get('Chad', 'N/A'))

my_dict = {'Ahmad': 1, 'Jane': 42}
my_dict.update({'John': 50})
print(my_dict)

my_dict = {'Ahmad': 1, 'Jane': 42}
val = my_dict.pop('Ahmad')
print(my_dict)
```

```python
# 
# Iterating over a dictionary

# 1. dict.items()

num_calories = dict(Coke=90, Coke_zero=0, Pepsi=94)

for soda, calories in num_calories.items():
    print(f'{soda}: {calories}')
```

```python
# 2. dict.keys()

for soda in num_calories.keys():
    print(soda)
```

```python


# 3. dict.values()

for calories in num_calories.values():
    print(calories)
```

```python
for key, value in enumerate(num_calories):
    print(key,' -- >',value)
```

```python
# Creating a nested dictionary
student_records = {"Jay": {"age": 20, "grade": "A", "subjects": ["Math","Science", "English"]},
"Ajay":{"age":19,"grade": "B","subjects": ["History","Geography","Physics"]},
"Bob": {"age": 21, "grade": "A", "subjects": ["Computer Science","Math","Chemistry"]}
}

print(student_records)
```

```python
for k,v in student_records.items():
    print(f'{k}:{v}')
```

```python
# Accessing data in the nested dictionary
print(student_records['Jay' ]['age' ]) # Output: 20



print(student_records["Ajay"]["subjects"] [0])
```
